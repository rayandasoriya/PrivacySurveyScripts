```python
import csv
import collections
import numpy as np
import matplotlib.pyplot as plt
```

```python
count = 0
data = []
cols = {}
with open("PrivacySheet1.csv") as queries:
    csv_reader = csv.reader(queries)
    next(csv_reader)
    count = 0
    for row in csv_reader:
        for i in range(19):
            if i in cols.keys():
                cols[i].append(row[i])
            else:
                cols[i] = [row[i]]
        data.append(row)
        count+=1
```

```python
# Implement the technology
health = collections.Counter(cols[5])
payment = collections.Counter(cols[9])
retail = collections.Counter(cols[13])

print(health, payment, retail)
```

```python
labels = list(health.keys())
sizes = list(health.values())
colors = ['lightskyblue','gold','lightcoral']

# Plot
plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Health: Implement')
plt.show()

labels = list(payment.keys())
sizes = list(payment.values())
colors = ['lightskyblue','gold','lightcoral']

# Plot
plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Payment: Implement')
plt.show()

labels = list(retail.keys())
sizes = list(retail.values())
colors = ['lightskyblue','gold','lightcoral']

# Plot
plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Retail: Implement')
plt.show()
```
