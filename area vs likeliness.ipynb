```python
import numpy as np
import time
import csv
import operator
import collections
```

```python
count = 0
data = []
cols = {}
with open("PrivacySheet.csv") as queries:
    csv_reader = csv.reader(queries)
    next(csv_reader)
    count = 0
    for row in csv_reader:
        data.append(row)
        for i in range(19):
            if i in cols.keys():
                cols[i].append(row[i])
            else:
                cols[i] = [row[i]]
        count+=1
```

```python

cols[18] = [x.lower() for x in cols[18]]
cols[18] = dict(collections.Counter(cols[18]))
print(len(cols[18]))
keys = list(cols[18].keys())
values = list(cols[18].values())
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects = list(keys)[:19]
y_pos = np.arange(len(objects))
performance = values

plt.barh(y_pos, performance, color='lightcoral', align='center', alpha=0.8)
plt.yticks(y_pos, objects)
plt.xlabel('Number of Suggestions')
plt.title('Suggested Domain by the Users')

plt.show()
```
