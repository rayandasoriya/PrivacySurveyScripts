```python
import csv
import collections
import numpy as np
import matplotlib.pyplot as plt
```

```python
count = 0
data = []
cols = {}
with open("PrivacySheet1.csv") as queries:
    csv_reader = csv.reader(queries)
    next(csv_reader)
    count = 0
    for row in csv_reader:
        for i in range(19):
            if i in cols.keys():
                cols[i].append(row[i])
            else:
                cols[i] = [row[i]]
        data.append(row)
        count+=1
```

```python
# FOR HEALTH -- Want to implement v/s Secure
a = collections.defaultdict(int)

for i in range(len(cols[4])):
    if cols[5][i] != 'No':
        if int(cols[7][i]) <=2:
            a['Less secure']+=1
        elif int(cols[7][i]) >3:
            a['More secure']+=1
        else:
            a['Moderately secure'] +=1
labels = list(a.keys())
sizes = list(a.values())
colors = ['lightskyblue','gold','lightcoral']

# Plot
plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Health: People who want to implement v/s security')
plt.show()
```
