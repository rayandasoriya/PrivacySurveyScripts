```python
import csv
import collections
import numpy as np
import matplotlib.pyplot as plt
```

```python
count = 0
data = []
cols = {}
with open("PrivacySheet1.csv") as queries:
    csv_reader = csv.reader(queries)
    next(csv_reader)
    count = 0
    for row in csv_reader:
        for i in range(19):
            if i in cols.keys():
                cols[i].append(row[i])
            else:
                cols[i] = [row[i]]
        data.append(row)
        count+=1
```

```python
# Like the technology
health = collections.defaultdict(int)
payment = collections.defaultdict(int)
retail = collections.defaultdict(int)
for i in range(len(cols[4])):
    if int(cols[4][i]) <= 2:
        health['Dislike'] += 1
    elif int(cols[4][i]) > 3:
        health['Like'] += 1
    else:
        health['Moderate'] += 1

    if int(cols[8][i]) <= 2:
        payment['Dislike'] += 1
    elif int(cols[8][i]) > 3:
        payment['Like'] += 1
    else:
        payment['Moderate'] += 1   
        
    if int(cols[12][i]) <= 2:
        retail['Dislike'] += 1
    elif int(cols[12][i]) > 3:
        retail['Like'] += 1
    else:
        retail['Moderate'] += 1

print(health, payment, retail)
```

```python
labels = list(health.keys())
sizes = list(health.values())
colors = ['lightskyblue','gold','lightcoral']

# Plot
plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Health: Likeness')
plt.show()

labels = list(payment.keys())
sizes = list(payment.values())
colors = ['lightskyblue','gold','lightcoral']

# Plot
plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Payment: Likeness')
plt.show()

labels = list(retail.keys())
sizes = list(retail.values())
colors = ['lightskyblue','gold','lightcoral']

# Plot
plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Retail: Likeness')
plt.show()
```
