```python
import csv
import collections
import numpy as np
import matplotlib.pyplot as plt
```

```python
count = 0
data = []
cols = {}
with open("PrivacySheet1.csv") as queries:
    csv_reader = csv.reader(queries)
    next(csv_reader)
    count = 0
    for row in csv_reader:
        for i in range(19):
            if i in cols.keys():
                cols[i].append(row[i])
            else:
                cols[i] = [row[i]]
        data.append(row)
        count+=1
print(collections.Counter(cols[3]))
```

```python
# Feature vs Privacy
health = collections.Counter(cols[6])
payment = collections.Counter(cols[10])
retail = collections.Counter(cols[14])
```

```python
labels = [i.split(',')[1] for i in list(health.keys())]
sizes = list(health.values())
colors = ['gold','lightcoral']

# Plot
plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Health: Feature vs Privacy')
plt.show()

labels = list(payment.keys())
sizes = list(payment.values())
colors = ['gold','lightcoral']

# Plot
plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Payment: Feature vs Privacy')
plt.show()

labels = list(retail.keys())
sizes = list(retail.values())
colors = ['gold','lightcoral']

# Plot
plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Retail: Feature vs Privacy')
plt.show()
```
